```python
import nltk
nltk.download('punkt')
```

```python
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Import NLTK libraries
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Step 3: Define the file path in Google Drive
file_path = '/content/drive/My Drive/Colab Notebooks/JaneEyre.txt'  # Update path as per your Drive structure

# Step 4: Read the file content
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()

# Step 5: Tokenize the text
tokens = word_tokenize(text)

# Step 6: Print the number of tokens
num_tokens = len(tokens)
print("Number of tokens:", num_tokens)

# Step 7: Identify the frequency of each token
frequency_dist = FreqDist(tokens)

# Display the frequency of the top 20 tokens
print("\nTop 20 most common tokens:")
for token, frequency in frequency_dist.most_common(20):
    print(f"{token}: {frequency}")
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from google.colab import drive
```

```python

# Step 3: Define the file path in Google Drive
file_path = '/content/drive/My Drive/Colab Notebooks/JaneEyre.txt'  # Update path as per your Drive structure

# Step 4: Read the file content
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    min_font_size=10
).generate(text)
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
from PIL import Image
import numpy as np

# Load the mask image
file_path = '/content/drive/My Drive/Colab Notebooks/wordcloud.jpg'  # Update with your mask path
mask = np.array(Image.open(file_path))

# Generate the word cloud with a mask
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    mask=mask,
    contour_width=1,
    contour_color="black"
).generate(text)

# Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
